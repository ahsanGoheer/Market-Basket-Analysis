```python
transaction_list=[]
for index,val in enumerate(groceries['Item(s)']):
    transaction_list.append(groceries.iloc[index,1:val+1].to_list())

    
         
```

```python
def apyori(df, minimum_support=0.1, confidence=0.22):
    df_values = df.values.astype(str)
    index, counts = np.unique(df_values,return_counts=True)
    df_item = pd.DataFrame(zip(index, counts), columns = ['produto', 'frequencia'])
    df_item.drop(df_item[(df_item['produto'] == 'nan' )|(df_item['produto'] == 'None' )].index, inplace=True)
    df_item.sort_values(by='frequencia', ascending=False, inplace=True)
    df_item.reset_index(drop=True, inplace=True)
    print(df_item)
```

```python
dataset=transaction_list
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
res=apriori(df,min_support=0.005,use_colnames=True).sort_values(by='support',inplace=False,ascending=False)
res['length']=res['itemsets'].apply(lambda x:len(x))
data_set=res.reset_index()
data_set[data_set['length']==data_set['length'].max()]
```

```python
arr=data_set[data_set['length']==data_set['length'].max()]
np.savetxt('top10_itemsets.txt',arr.values.astype(str),fmt="%s",delimiter=',')
```

```python
from mlxtend.frequent_patterns import association_rules
a_rules=association_rules(res.iloc[:,0:2],support_only=False,min_threshold=0.08)
a_rules
```

```python
plt.scatter(a_rules.support,a_rules.confidence,c=a_rules.lift,cmap='gray')
```

```python
min_support_threshold=700
min_confidence=0.5

#Get item frequencies.
columns={'index':'Item',0:'Frequency'}
item_freq=item_counts.rename(columns=columns,inplace=False)

#Applying Apriori Algorithm

frequent_items=item_freq[item_freq['Frequency']>=min_support_threshold]

frequent_itemsets=pd.DataFrame(columns=['itemset','frequency'])

for n in range(1,len(frequent_items)+1):
    total_combinations= list(combinations(frequent_items['Item'].values,n))
    for combination in total_combinations:
        counter=0
        for transaction in transaction_list:
            if all(item in transaction for item in combination):
                counter=counter+1
        if counter>= min_support_threshold:
            frequent_itemsets= frequent_itemsets.append({'itemset':combination,'frequency':counter},ignore_index=True)
    frequent_itemsets.sort_values(by='frequency',inplace=True,ascending=False)
    
print(frequent_itemsets)

```
